```python
import numpy as np
```

```python
%%time
def calculate_hotcoldmass(Masses, Temperature, SubhaloID, parttype="PartType0",shape=(2,)):
    """Return gas mass above and below a temperature threshold Tthresh for the central subhalo."""
    Tthresh = 1e6 # K
    #mask = SubhaloID==0
    #if np.count_nonzero(mask)==0:
    #    return np.zeros(2)
    hotgas = np.where(Temperature>=Tthresh, Masses, 0).sum()
    coldgas = np.where(Temperature<Tthresh, Masses, 0).sum()
    return np.array([hotgas,coldgas])

#masses = snap.map_halo_operation(calculate_hotcoldmass, chunksize=int(3e9), compute=True)
masses = ds.map_group_operation(calculate_hotcoldmass, cpucost_halo=10000.0, nchunks_min=None, chunksize_bytes=int(3e9), idxlist=None, objtype='halo', compute=True)
```

```python
%%time
# Get Halo Mass to plot against
masses_h = ds.data["Group"]["GroupMass"].compute()
print(np.shape(masses_h))
```

```python
from astropy import units as u
from astropy.cosmology import FlatLambdaCDM
```

```python
cosmo = FlatLambdaCDM(H0=67.74, Om0=0.3089)
littleh = cosmo.H0.value/100.
Mpcph = u.def_unit('Mpcph', u.Mpc/littleh) #[Mpc/h]
Msunph = u.def_unit('Msunph', u.Msun/littleh) #[Msun/h]
kpcph = u.def_unit('kpcph', u.kpc/littleh) #[kpc/h]
```

```python
R22_0p015_kpc = (((R22_0p015.magnitude) * kpcph ).to(u.kpc)).value
```

```python
r_half_star_kpc = ((r_half_star.magnitude * kpcph ).to(u.kpc)).value
```

```python
r_half_gas_kpc = ((r_half_gas.magnitude * kpcph ).to(u.kpc)).value
```

```python
r_half_gas_kpc = ((r_half_gas.magnitude * kpcph ).to(u.kpc)).value
```
